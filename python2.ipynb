```python
import matplotlib.pyplot as plt
import pandas as pd

```

```python
df = pd.read_csv('D:/data/movies.csv')
print(df.head())
```

```python
df.info()
```

```python
df.head()
```

# Top 10 movies by box office

```python
top10 = df.sort_values(by='box_office_million',ascending=False).head(10)
plt.figure(figsize=(12,5))
plt.bar(top10['title'],top10['box_office_million'])
plt.show()
```

# Number of movies per genre

```python
gen_count = df.groupby('genre')['movie_id'].count()
plt.figure(figsize=(8,4))
plt.bar(gen_count.index,gen_count.values)
plt.show()
```

# IMDb rating by platform

```python
imd = df.groupby('platform')['imdb_rating'].mean().round(2)
plt.figure(figsize=(8,5))
plt.bar(imd.index,imd.values)
plt.show()
```

# Box office collection trend over years

```python
coll = df.groupby ('release_year')['box_office_million'].sum()
plt.figure(figsize=(8,3))
plt.plot(coll.index,coll.values)
plt.show()
```

# Rating trend across years

```python
rating = df.groupby('release_year')['imdb_rating'].mean().round(2)
plt.figure(figsize=(8,3))
plt.plot(rating.index,rating.values)
plt.show()
```

# Genre distribution

```python
dis = df.groupby('genre')['movie_id'].count()
plt.figure(figsize=(4,4))
plt.pie(dis.values,labels = dis.index,autopct='%1.1f%%')

plt.show()
```

# Platform distribution 

```python
plat = df.groupby('platform')['movie_id'].count()
plt.figure(figsize=(5,5))
plt.pie(plat.values , labels=plat.index,autopct='%1.1f%%')
plt.show()
```

# Budget vs Box Office

```python
plt.figure(figsize=(5,5))
plt.scatter(df['budget_million'],df['box_office_million'],cmap='viridis')
plt.colorbar()
plt.show()
```

# Duration vs IMDb Rating

```python
plt.figure(figsize=(5,5))
plt.scatter(df['duration_min'],df['imdb_rating'],cmap='viridis')
plt.colorbar()
plt.show()
```

```python
plt.hist(df['duration_min'],edgecolor='r')
plt.show()
```
